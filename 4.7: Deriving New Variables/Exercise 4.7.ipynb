```python
# user defined function for price ranges

def price_label(row):
    if row['prices'] <= 5:
        return 'Low range products'
    elif row ['prices'] > 5 and row ['prices'] <= 15:
        return 'Middle range products'
    elif row ['prices'] > 15:
        return 'Expensive Products'
    else: 
        return 'Not available'
```

```python
#creating a new column 'price_range'
df_ords_prods_subset['price_range'] = df_ords_prods_subset.apply(price_label , axis =1)
```

```python
df_ords_prods_subset['prices'].max()
```

```python
#checking the frequency of column : 'price_range'
df_ords_prods_subset['price_range'].value_counts(dropna = True)
```

```python
#filtering using loc function
df_ords_prods_subset.loc[df_ords_prods_subset['prices'] <= 5 ,'prices_range_loc'] = 'Low end products'
df_ords_prods_subset.loc[(df_ords_prods_subset['prices'] > 5) & (df_ords_prods_subset['prices'] <= 15 ),'prices_range_loc'] = 'Mid range products'
df_ords_prods_subset.loc[df_ords_prods_subset['prices'] > 15 , 'prices_range_loc'] = "High end products"
```

```python
#Update "busiest day" column to "busiest days" to identify order day of the week as "Busiest days," "Slowest days," or "Regularly busy."
result_2 = []

for i in df_ords_prods_merg['order_day']:
    if i == 0 or i == 1:
        result_2.append("Busiests Days")
    elif i ==3 or i == 4:
        result_2.append("Slowest Days")
    else:
        result_2.append("Regular Days")
```

```python
#creating new column 'Busiest Days'
df_ords_prods_merg['Busiest Days'] = result_2
```

```python
#checking frequency of 'Busiest Days' column
df_ords_prods_merg['Busiest Days'].value_counts(dropna = True)
```

# identify the busiest hours of the day

```python
df_ords_prods_merg['order_hour_of_day'].value_counts(dropna = True)
```

The value counts listed above will be split into equal thirds for the following labels: "Most orders" = 10, 11, 14, 15, 13, 12, 16, and 9; "Fewest orders" = 23, 6, 0, 1, 5, 2, 4, and 3; "Average orders" = all remaining values in order_hour_of_day.

```python
result_3 = []

for values in df_ords_prods_merg['order_hour_of_day']:
    if values in [10, 11, 14, 15, 13, 12, 16,9] :
        result_3.append("Most orders")
    elif values in [23, 6, 0, 1, 5, 2, 4,3]:
        result_3.append("Fewest orders")
    else:
        result_3.append("Average orders")
        
    
    
```

```python
df_ords_prods_merg['busiest_period_of_day'] = result_3
```

```python
df_ords_prods_merg['busiest_period_of_day'].value_counts(dropna = True)
```

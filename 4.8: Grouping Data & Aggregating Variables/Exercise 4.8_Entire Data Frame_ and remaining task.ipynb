# Grouping using pandas

```python
#aggregating data with agg()
df_ords_prods_merge_updated.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
#other way
df_ords_prods_merge_updated.groupby('department_id')['order_number'].mean()
```

```python
#performing multiple aggregtion
df_ords_prods_merge_updated.groupby('department_id').agg({'order_number' :['mean','max','min']})
```

```python
#just for practice
def order (row):
    if row('order_number') > 40:
        return 'Loyal Cutomer'
    elif row ('order_number') <= 40 and row('order_number') > 10:
        return 'Regular customer'
    elif row('order_number') <= 10:
        return 'New Customer'
    else:
        return 'Not available'
```

```python
df_ords_prods_merge_updated['max_order'] = df_ords_prods_merge_updated.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge_updated.loc[(df_ords_prods_merge_updated['max_order'] <= 40) & (df_ords_prods_merge_updated['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge_updated['loyalty_flag'].value_counts()
```

```python
df_ords_prods_merge_updated.groupby('loyalty_flag').agg({'prices' :['mean','max','min']})
```

```python
#create a new column average price
df_ords_prods_merge_updated['average_price'] = df_ords_prods_merge_updated.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge_updated[['user_id','average_price']]
```

```python
#creating a spending flag
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['average_price'] < 10, 'Spending Type'] = 'Low Spender'
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['average_price'] >= 10, 'Spending Type'] = 'Higher Spender'
```

```python
#counting frequency of spending flag
df_ords_prods_merge_updated['Spending Type'].value_counts(dropna =True)
```

```python
df_ords_prods_merge_updated[['user_id','average_price','Spending Type','prices']].head(10)
```

```python
#creating a column of median of days prior order
df_ords_prods_merge_updated['Median of days prior order'] = df_ords_prods_merge_updated.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#frequency of the median days prior order
df_ords_prods_merge_updated['Median of days prior order'].value_counts(dropna =False)
```

```python
df_ords_prods_merge_updated[['user_id','Median of days prior order','days_since_prior_order']].head(20)
```

```python
#creating a frequent customer flag
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['Median of days prior order'] > 20, 'Customer Frequency'] = 'Non-frequent customer'
df_ords_prods_merge_updated.loc[(df_ords_prods_merge_updated['Median of days prior order'] > 10) & (df_ords_prods_merge_updated['Median of days prior order'] <= 20), 'Customer Frequency'] = 'Regular customer'
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['Median of days prior order'] <= 10, 'Customer Frequency'] = 'Frequent customer'
```

```python
#counting the frequency of frequent customer flag
df_ords_prods_merge_updated['Customer Frequency'].value_counts(dropna = False)
```

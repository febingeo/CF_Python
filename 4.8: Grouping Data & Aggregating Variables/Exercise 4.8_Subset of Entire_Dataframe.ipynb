## Grouping using pandas for subset 

```python
    df_ords_prods_merge_updated_subset.groupby('product_name')
```

```python
#aggregating data with agg() for mean
df_ords_prods_merge_updated_subset.groupby('department_id').agg({'order_number':['mean']})
```

```python
#other way
df_ords_prods_merge_updated_subset.groupby('department_id')['order_number'].mean()
```

```python
#performing multiple aggregtion
df_ords_prods_merge_updated_subset.groupby('department_id').agg({'order_number': ['mean','max','min']})
```

```python
#Aggregating Data with transform()
df_ords_prods_merge_updated_subset['max_order'] = df_ords_prods_merge_updated_subset.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge_updated_subset
```

```python
#creating the loyalty flag with loc[] function
df_ords_prods_merge_updated_subset.loc[df_ords_prods_merge_updated_subset['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge_updated_subset.loc[(df_ords_prods_merge_updated_subset['max_order'] <= 40) & (df_ords_prods_merge_updated_subset['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge_updated_subset.loc[df_ords_prods_merge_updated_subset['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
